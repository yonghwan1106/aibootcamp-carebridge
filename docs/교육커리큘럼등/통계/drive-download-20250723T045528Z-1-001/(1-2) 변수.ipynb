**3. print() 함수**

```python
# 문자열 출력
print("오영석")

# 변수 출력
a = 10
print(a)

# 계산 결과 출력
print(10+20)
print("I" + "Love" + "You")

# 여러 값을 한 번에 출력
print("오영석", "Orange", "3")
```

```python
# 기본 공백으로 구분
print("오영석", "Orange", "3")
```

```python
# 쉼표로 구분
print("오영석", "Orange", "3", sep= ", ")
```

```python
# 파이프로 구분
print("오영석", "Orange", "3", sep="|")
```

```python
# 줄바꿈으로 구분
print("오영석", "Orange", "3", sep="\n")
```
