**2 벡터의 표현**

```python
b = np.array([175, 60, 81, 0.8, 0.9])
print(b)
```

```python
# 벡터의 스칼라곱 코드 표현
b = np.array([175, 60, 81, 0.8, 0.9])
c = 3

print(b*c)
```

```python
# 벡터의 합과 차 코드 표현
b = np.array([175, 60, 81, 0.8, 0.9])
d = np.array([180, 70, 86, 1.2, 1.5])

print(b+d)
```

```python
print(d-b)
```

```python
# 벡터의 성분곱 코드 표현
b = np.array([175, 60, 81, 0.8, 0.9])
d = np.array([180, 70, 86, 1.2, 1.5])

print(b*d)
```

**3. 행렬의 표현**

```python
A = np.array([[1, 3, 4],
              [3, 2, 1]])
B = np.array([[2, 4, 3],
              [1, 3, 2]])
print(A+B)
```

```python
print(A-B)
```
