**1. 리스트 컴프리헨션의 코드 표현**

```python
# 기존의 방법 (3줄)
numbers = []
for i in range(10):
    numbers.append(i * 2)
print(numbers)
```

```python
# 리스트 컴프리헨션 (1줄)
numbers = [i * 2 for i in range(10)]
print(numbers)
```

```python
# 원본 데이터
raw_scores = [85, 92, 78, 95, 88, 72, 96, 83]

# 기존의 방법으로 0-1 범위로 정규화
max_score = max(raw_scores)
min_score = min(raw_scores)
normalized_traditional = []
for score in raw_scores:
    normalized_score = (score - min_score) / (max_score - min_score)
    normalized_traditional.append(normalized_score)

print(normalized_traditional)
```

```python
# 리스트 컴프리헨션으로 0-1 범위로 정규화
normalized = [(score - min_score) / (max_score - min_score) for score in raw_scores]
print(normalized)
```

```python
# 반복문과 슬라이스 기능을 이용한 리스트 컴프리헨션

raw_data = [[1, 10], [2, 21], [3, 32], [4, 46]]

all_data = [x for x in raw_data]
x_data = [x[0] for x in raw_data]
y_data = [x[1] for x in raw_data]
```

```python
print(all_data)
print(x_data)
print(y_data)
```

**2. while 반복문의 코드 표현**

```python
count = 1
while count <= 5:
    print(f"카운트: {count}")
    count += 1  # 조건을 변화시키는 부분
```

```python
# For 반복문과의 차이
for count in range(1, 6):
    print(f"카운트: {count}")
```

```python
numbers = [10, 20, 30, 40, 50, 60]
i = 0
while i < len(numbers):
    print(f"인덱스 {i}: {numbers[i]}")
    i += 1
```
