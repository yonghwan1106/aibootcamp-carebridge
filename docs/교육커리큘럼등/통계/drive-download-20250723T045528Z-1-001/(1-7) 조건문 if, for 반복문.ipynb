**1 조건문 if의 코드 표현**

```python
# if 조건식만을 사용하는 방식

p = 7
if p > 0:
  print("p ==", p)
  print("양의 정수")

elif p == 0:
  print("p ==", p)
  print("영")

else:
  print("p ==", p)
  print("음의 정수")
```

```python
# if 조건식과 함께 in 키워드를 사용하는 방식

list_data = [10, 20, 30, 40, 50, 60]

if 50 in  list_data:
  print("50은 list_data에 포함되어 있습니다.")

else:
  print("50은 list_data에 포함되어 있지 않습니다.")
```

```python
dic_data = {"Kim": 90, "Lee": 85, "Oh": 95}

if "Kim" in dic_data:
  print("Kim은 dic_data에 포함되어 있습니다.")

else:
  print("Kim은 dic_data에 포함되어 있지 않습니다.")
```

**2 for 반복문의 코드 표현**

```python
# in range(..) 형태를 사용하는 방식

for data in range(1, 11):
  print(data, " ")
```

```python
for data in range(1, 11):
  print(data, " ", end = "")
```

```python
# for문에서 in 다음에 리스트 또는 딕셔너리 등의 데이터 타입을 표시하는 방식

list_data = [10, 20, 30, 40, 50, 60]
for data in list_data:
  print(data, " ", end="")
```

```python
dic_data = {"Kim": 90, "Lee": 85, "Oh": 95}
for key, value in dic_data.items():
  print(key, " ", value)
```

```python
dic_data = {"Kim": 90, "Lee": 85, "Oh": 95}
for key in dic_data.keys():
  print(key, " ")
```

```python
dic_data = {"Kim": 90, "Lee": 85, "Oh": 95}
for value in dic_data.values():
  print(value, " ")
```
