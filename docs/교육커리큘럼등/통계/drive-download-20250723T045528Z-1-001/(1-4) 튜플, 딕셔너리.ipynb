**2 딕셔너리의 코드 표현**

```python
dic = {"Kim": 90, "Lee": 85, "Oh": 95 }
print(dic)
```

```python
dic["Kim"] = 93
print(dic)
```

```python
dic["Choi"] = 100
print(dic)
```

```python
dic[3]
```

```python
dic = {"Kim": 90, "Lee": 85, "Oh": 95, ["Song", "Choi"]: 80}
print(dic)
```

```python
dic = {"Kim": 90, "Lee": 85, "Oh": 95, ("Song", "Choi"): 80}
print(dic)
```

```python
del dic["Oh"]
print(dic)
```

```python
print(dic.keys())
print(dic.values())
print(dic.items())
```
