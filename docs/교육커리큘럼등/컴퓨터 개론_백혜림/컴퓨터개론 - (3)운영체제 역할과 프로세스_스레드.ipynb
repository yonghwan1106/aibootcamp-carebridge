 ## 🖥️⚙️👑 운영체제 (Operating System, OS) 란?



 * **정의**:

   컴퓨터 **하드웨어**와 **사용자(또는 애플리케이션 소프트웨어)** 사이에 위치하여,

   컴퓨터 시스템의 **자원(Resource)** 을 효율적으로 **관리**하고

   사용자에게 **편리한 인터페이스**를 제공하는 핵심 **시스템 소프트웨어**.

 * **역할**:

   컴퓨터 시스템의 **관리자**이자 **중재자**.

   하드웨어의 복잡성을 숨기고 사용자가 컴퓨터를 쉽게 사용할 수 있도록 도와줌.



 * **비유**:

     * **오케스트라 지휘자**:

       다양한 악기 연주자(하드웨어 장치)들을 조율하고 지휘하여

       아름다운 음악(애플리케이션 실행)을 만들어내는 역할. 🎶



 * **운영체제의 핵심 기능**:

     1.  **프로세스 관리 (Process Management)**: 프로그램의 생성, 실행, 종료, 스케줄링, 동기화 등 관리. (오늘 핵심 주제!)

     2.  **메모리 관리 (Memory Management)**: 프로그램 실행에 필요한 메모리 공간 할당 및 회수, 가상 메모리 관리 등. (2교시 연관)

     3.  **저장장치 관리 (Storage Management / File System)**: 디스크 공간 관리, 파일 생성/삭제/읽기/쓰기 등 파일 시스템 운영. (4교시 연관)

     4.  **입출력 장치 관리 (I/O Device Management)**: 키보드, 마우스, 모니터, 프린터 등 주변 장치와의 데이터 입출력 제어.

     5.  **네트워킹 (Networking)**: 네트워크 통신 기능 제공 및 관리. (5일차 수업 연관)

     6.  **보안 (Security)**: 시스템 접근 제어, 사용자 인증, 자원 보호 등.

     7.  **사용자 인터페이스 (User Interface)**: 사용자가 컴퓨터와 상호작용할 수 있는 환경 제공 (CLI - Command Line Interface, GUI - Graphical User Interface).



 * **대표적인 운영체제**: Windows, macOS, **Linux**, Unix, iOS, Android 등.

 ## 👨‍🍳 vs 👯‍♀️ 프로세스 (Process) 와 스레드 (Thread)



 운영체제가 관리하는 작업의 기본 단위들입니다.



 ### 1. 프로세스 (Process)



 * **정의**: **실행 중인 프로그램(Program in Execution)**. 디스크에 저장된 프로그램 코드가 메모리에 올라와 CPU에 의해 실행되는 **동적인 상태**.

 * **특징**:

     * **독립된 메모리 공간**: 각 프로세스는 자신만의 **독립적인 주소 공간**(코드, 데이터, 힙, 스택 영역)을 할당받음. 다른 프로세스의 메모리 공간에 직접 접근 불가 (OS의 보호). 🛡️

     * **자원 소유 단위**: 운영체제로부터 CPU 시간, 메모리, 파일 핸들 등 자원을 할당받는 단위.

     * **프로세스 간 통신 (IPC - Inter-Process Communication)**: 독립된 메모리 때문에, 프로세스끼리 데이터를 주고받으려면 별도의 IPC 메커니즘(파이프, 소켓, 공유 메모리 등) 필요. 복잡함. 🔴

 * **구성 요소 (메모리 구조)**:

     * **코드(Code) 영역**: 실행할 프로그램 명령어.

     * **데이터(Data) 영역**: 전역 변수, 정적 변수 등 저장.

     * **힙(Heap) 영역**: 프로그램 실행 중 동적으로 할당되는 메모리 공간 (예: 객체 생성).

     * **스택(Stack) 영역**: 함수 호출 시 지역 변수, 매개변수, 복귀 주소 등 저장.

 * **비유**: 하나의 **요리 레시피(프로그램)** 를 가지고 **독립된 주방(메모리 공간)** 에서 요리를 하고 있는 **셰프(프로세스)** 👨‍🍳. 각 셰프는 자신만의 재료와 도구를 사용.

 ## 1. 데이터 세그먼트 (Data Segment) 영역



 프로그램이 시작될 때 할당되어 종료될 때까지 유지되는 **전역 변수(Global variables)**와 **정적 변수(Static variables)**가 저장되는 영역입니다.



 * **초기화된 데이터 (Initialized Data):** 코드에서 초기값을 지정한 전역/정적 변수 (예: `global_var_int = 100`)

 * **초기화되지 않은 데이터 (BSS - Block Started by Symbol):** 코드에서 초기값을 지정하지 않은 전역/정적 변수 (C에서는 0으로 초기화, 파이썬에서는 `None` 등으로 표현하거나 사용 전에 할당해야 함)



 파이썬에서는 모든 것이 객체이므로, 실제 객체(정수, 문자열, 리스트 등)는 대부분 **힙(Heap)** 영역에 생성되고, 데이터 세그먼트 영역에는 이 객체를 가리키는 **참조(이름)**가 저장된다고 개념적으로 이해할 수 있습니다.

````python
# %%
# --- 데이터 세그먼트 (Data Segment) 영역 ---

global_var_int = 100  # 초기화된 데이터 (Initialized Data)
global_var_str = "Hello, Global Data"
global_var_list = [1, 2, 3] # 리스트 객체 자체는 힙에 생성, 이 변수는 그 객체를 가리킴

# 초기화되지 않은 데이터(BSS)는 파이썬에서 명시적으로 구분하기 어려움 (보통 0이나 None으로 초기화됨)
global_uninitialized = None # 예시

print("--- [데이터 세그먼트 영역 (전역 변수)] ---")
print(f"전역 정수 변수     (global_var_int): 값={global_var_int}, ID(메모리 주소 추정)={id(global_var_int)}")

# 작은 정수 (-5 ~ 256)는 파이썬 내부적으로 캐싱(interning)되어 동일 ID를 가질 수 있습니다.
# 즉, 같은 값을 가진 작은 정수는 메모리 상에 오직 하나만 존재할 수 있습니다.
small_int_1 = 100
small_int_2 = 100
print(f"작은 정수 100 (small_int_1) ID 확인: {id(small_int_1)}")
print(f"작은 정수 100 (small_int_2) ID 확인: {id(small_int_2)} (global_var_int와 ID 동일 가능성 높음)")

large_global_int1 = 999999999 # 큰 정수는 별도 객체로 생성될 가능성 높음
large_global_int2 = 999999999
print(f"큰 전역 정수 변수 (large_global_int1): 값={large_global_int1}, ID={id(large_global_int1)}")
print(f"큰 전역 정수 변수 (large_global_int2): 값={large_global_int2}, ID={id(large_global_int2)}")

print(f"전역 문자열 변수   (global_var_str): 값='{global_var_str}', ID={id(global_var_str)}")
# 짧은 문자열도 캐싱(interning)될 수 있습니다.
short_str_1 = "abc"
short_str_2 = "abc"
print(f"짧은 문자열 'abc' (short_str_1) ID: {id(short_str_1)}")
print(f"짧은 문자열 'abc' (short_str_2) ID: {id(short_str_2)} (ID 동일 가능성 높음)")

long_str_1 = "Hello, Global Data"
long_str_2 = "Hello, Global Data"
print(f"긴 문자열 (long_str_1) ID: {id(long_str_1)}")
print(f"긴 문자열 (long_str_2) ID: {id(long_str_2)} (ID 다름)")

print(f"전역 리스트 변수   (global_var_list): 값={global_var_list}, ID={id(global_var_list)}")
print(f"  ㄴ 리스트 객체 내부 첫 요소(1) ID: {id(global_var_list[0])} (정수 1은 캐싱됨)")
print(f"초기화 안된 전역 변수(global_uninitialized): 값={global_uninitialized}, ID={id(global_uninitialized)}")
````

 ## 2. 코드 세그먼트 (Code/Text Segment) 영역



 실행 가능한 **기계 코드(명령어)**가 저장되는 영역입니다. 일반적으로 읽기 전용(Read-Only)이며, 프로그램의 동작 로직을 담고 있습니다.



 파이썬에서는 인터프리터 자체의 코드와 우리가 작성한 파이썬 코드가 컴파일된 **바이트코드(Bytecode)**가 이 영역에 해당한다고 볼 수 있습니다.



 함수 정의(def)나 클래스 정의(class) 등이 코드 세그먼트의 일부입니다.

````python
import dis

# %%
# --- 코드 세그먼트 (Code/Text Segment) 영역 ---

def sample_function(param1, param2):
    """
    이 함수 정의 자체가 코드 세그먼트에 위치하는 코드의 일부입니다.
    함수 호출 시, 이 함수의 바이트코드가 실행됩니다.
    """
    # 함수 내부 로직은 아래 스택/힙 설명에서 실행됩니다.
    local_var_inside_func = param1 + 10
    print(f"    [sample_function 내부] 지역 변수 local_var_inside_func: 값={local_var_inside_func}, ID={id(local_var_inside_func)}")
    return local_var_inside_func

# 함수 객체 자체의 ID 확인 (코드 세그먼트 내 위치 추정)
print(f"--- [코드 세그먼트 영역 (함수 객체)] ---")
print(f"sample_function 객체 자체의 ID: {id(sample_function)}")
print(dis.dis(sample_function))

# inspect 모듈로 함수의 바이트코드 정보 일부 확인 가능 (참고)
# print("\n--- 함수의 바이트코드 정보 (일부) ---")
# print(inspect.getsource(sample_function)) # 소스코드 출력
# print(sample_function.__code__.co_code) # 실제 바이트코드 (bytes 객체)
````

 ## 3. 스택(Stack) & 힙(Heap) 영역: 함수 호출과 객체 생성



 ### 스택 영역 (Stack Segment)



 함수가 호출될 때 **지역 변수(Local variables)**, **매개변수(Parameters)**,

 함수가 끝나고 돌아갈 주소 등이 저장되는 공간입니다.



 함수가 시작되면 그 함수를 위한 **스택 프레임(Stack Frame)**이 만들어져

 스택에 쌓이고(push), 함수가 끝나면 스택 프레임이 제거됩니다(pop).



 메모리를 아주 빠르게 할당하고 해제할 수 있지만,

 사용할 수 있는 공간이 제한되어 있습니다. (Last-In, First-Out 구조)

 ### 힙 영역 (Heap Segment)



 프로그램이 실행되는 동안 **필요할 때마다 메모리를 할당**하고 해제하는 공간입니다.



 사용자가 원하는 때(예: 객체를 만들 때) 메모리를 할당받고,

 더 이상 필요 없어지면(참조 카운트가 0이 되면)

 가비지 컬렉터(Garbage Collector)가 메모리를 해제합니다.



 스택보다는 속도가 느리지만,

 훨씬 더 큰 메모리 공간을 사용할 수 있습니다.

 **파이썬에서는 대부분의 객체(리스트, 딕셔너리, 사용자 정의 클래스 인스턴스 등)가

 힙 영역에 생성됩니다.**



 스택 영역에는 이 힙 영역에 생성된 객체를 가리키는 **참조(이름)**가

 저장되는 경우가 많습니다.

````python
# %%
# --- 스택(Stack) 및 힙(Heap) 영역 시연 ---

print("\n--- [함수 호출 및 스택/힙 사용] ---")

def stack_heap_demo(param_x, param_y):
    """
    함수 호출 시 스택 프레임이 생성되고, 지역 변수가 할당됩니다.
    함수 내에서 생성되는 객체는 주로 힙에 할당됩니다.
    """
    print(f"--- 함수 진입: {stack_heap_demo.__name__} (스택 프레임 생성) ---")
    print(f"  매개변수 param_x (값: {param_x}) - ID: {id(param_x)} (스택에 참조 저장)")
    print(f"  매개변수 param_y (값: '{param_y}') - ID: {id(param_y)} (스택에 참조 저장)")

    # --- Stack (사례: 지역 변수) ---
    local_var_int = 200 # 값 자체 또는 참조가 스택에 저장
    local_var_str = "Hello, Stack" # 문자열 객체는 힙, 참조는 스택
    local_var_list = [4, 5, 6] # 리스트 객체는 힙, 참조는 스택

    print("\n  --- 지역 변수 (스택 프레임 내) ---")
    print(f"  local_var_int   (값: {local_var_int}) - ID: {id(local_var_int)}")
    # 작은 정수 캐싱 주의
    large_local_int = 9876543210
    print(f"  large_local_int (값: {large_local_int}) - ID: {id(large_local_int)}")
    print(f"  local_var_str   (값: '{local_var_str}') - ID: {id(local_var_str)}")
    print(f"  local_var_list  (값: {local_var_list}) - ID: {id(local_var_list)}")

    # --- Heap (사례: 함수 내 동적 객체 생성) ---
    heap_object_dict = {'key': 'value', 'ref': local_var_list}
    heap_object_set = {1, 2, 3, 3} # Set 객체도 힙에 생성

    print("\n  --- 함수 내 동적 생성 객체 (힙) ---")
    print(f"  heap_object_dict (값: {heap_object_dict}) - ID: {id(heap_object_dict)}")
    print(f"    ㄴ 딕셔너리가 참조하는 local_var_list ID: {id(heap_object_dict['ref'])} (위 local_var_list ID와 동일)")
    print(f"  heap_object_set  (값: {heap_object_set}) - ID: {id(heap_object_set)}")

    # 클래스 인스턴스 생성 역시 힙에 할당됩니다.
    class MyClass:
        def __init__(self, name):
            self.name = name # 인스턴스 변수도 힙에 있는 인스턴스 내부에 저장

    instance_obj = MyClass("HeapInstance")
    print(f"  MyClass 인스턴스 (값: {instance_obj}) - ID: {id(instance_obj)}")
    print(f"    ㄴ 인스턴스 변수 name ('{instance_obj.name}') ID: {id(instance_obj.name)}")

    print(f"--- 함수 종료: {stack_heap_demo.__name__} (스택 프레임 소멸 임박) ---")
    # 함수가 반환되면 지역 변수(local_var_int, local_var_str, local_var_list 등) 이름은 사라집니다.
    # 힙에 생성된 객체(리스트 [4,5,6], 딕셔너리, 셋, MyClass 인스턴스)는
    # 더 이상 참조되지 않으면 (참조 카운트 0) 가비지 컬렉터 대상이 됩니다.
    return large_local_int # 값을 반환

# 함수 호출
print("=== 함수 호출 시작 ===")
arg1_val = 10
arg2_val = "Argument String"
print(f"호출 전 arg1_val ID: {id(arg1_val)}")
print(f"호출 전 arg2_val ID: {id(arg2_val)}")

returned_value = stack_heap_demo(arg1_val, arg2_val)

print("\n=== 함수 호출 완료 ===")
print(f"함수로부터 반환된 값: {returned_value}, ID: {id(returned_value)}")

# 함수 호출 후 지역 변수 접근 시도 (오류 발생 확인)
try:
    print(local_var_int)
except NameError as e:
    print(f"\n함수 종료 후 지역 변수 접근 시도 결과: NameError - {e}")
````

 ## 4. 스택(Stack) 영역의 변화: 재귀 호출



 재귀 함수(자기 자신을 호출하는 함수)를 사용하면 함수가 호출될 때마다 스택 프레임이 계속해서 쌓이는 것을 관찰할 수 있습니다. 너무 깊게 재귀 호출을 하면 스택 공간이 부족해져 **스택 오버플로우(Stack Overflow)** 오류가 발생합니다.

````python
# %%
# --- 스택 영역의 변화 (재귀 호출) ---

def recursive_stack_demo(depth):
    """ 재귀 호출을 통해 스택이 깊어지는 것을 보여줍니다. """
    print(f"  재귀 깊이 {depth}, 스택 변수(depth) ID: {id(depth)}")
    # 각 호출마다 depth 변수는 해당 스택 프레임에 저장됩니다. (작은 정수 캐싱 주의)
    local_data_at_depth = f"Data at depth {depth}"
    print(f"  재귀 깊이 {depth}, 지역 변수(local_data_at_depth) ID: {id(local_data_at_depth)}") # 매번 다른 문자열 객체 생성 (힙), 참조는 스택

    if depth > 0:
        recursive_stack_demo(depth - 1) # 자기 자신 호출
    else:
        print("  --- 재귀 호출 가장 깊은 지점 도달 ---")

    # 함수가 반환되면서 스택 프레임이 하나씩 해제됨
    print(f"  --- 재귀 반환: 깊이 {depth} 프레임 해제 임박 ---")

print("\n--- [재귀 함수 호출로 스택 변화 관찰] ---")
# 너무 깊은 재귀는 RecursionError (Stack Overflow)를 유발할 수 있으므로 작은 깊이로 테스트합니다.
try:
    recursive_stack_demo(3) # 깊이 3으로 호출
except RecursionError as e:
    print(f"재귀 호출 중 오류 발생: {e}")
print("--- [재귀 함수 호출 완료] ---")
````

 ### 2. 스레드 (Thread)



 * **정의**: **프로세스 내**에서 실행되는 **실행 흐름의 가장 작은 단위**. "경량 프로세스(Lightweight Process)"라고도 불림.

 * **특징**:

     * **프로세스 자원 공유**: 스레드는 자신이 속한 프로세스의 **코드, 데이터, 힙 영역 메모리를 다른 스레드들과 공유**함! 👍🔵

     * **독립적인 실행 흐름**: 각 스레드는 자신만의 **독립적인 스택(Stack) 영역**과 **레지스터(Register)** 값을 가짐. (함수 호출, 지역 변수 등은 독립적)

     * **스레드 간 통신**: 같은 프로세스 내 스레드들은 메모리를 공유하므로, **데이터를 직접 주고받기 쉬움**. (전역 변수, 공유 객체 등 사용) ✅🔵

     * **생성/관리 오버헤드 적음**: 프로세스 생성보다 스레드 생성이 훨씬 가볍고 빠름.

 * **비유**: **같은 주방(프로세스 메모리)** 에서 **같은 레시피(코드)** 와 **공유된 재료(데이터, 힙)** 를 사용하며 각자 다른 작업(예: 채소 썰기, 소스 만들기)을 하는 **여러 명의 요리사(스레드)** 👯‍♀️. 각자 작업대(스택)는 따로 사용.

 ## 🚀<🚀 멀티프로세싱 vs 멀티스레딩



 하나의 컴퓨터에서 여러 작업을 동시에 처리하는 방식.



 * **멀티프로세싱 (Multiprocessing)**: **여러 개의 프로세스**를 동시에 실행하는 방식.

     * **장점**:

         * **진정한 병렬 처리 (True Parallelism)**: 멀티 코어 CPU 환경에서 각 코어가 다른 프로세스를 동시에 실행하여 성능 극대화 가능. 🚀

         * **안정성**: 한 프로세스에 문제가 생겨도 다른 프로세스에 영향을 주지 않음. 👍

     * **단점**:

         * **높은 오버헤드**: 각 프로세스는 독립적인 메모리를 가지므로 생성 및 컨텍스트 스위칭(작업 전환) 비용이 큼. 🐢🔴

         * **복잡한 통신 (IPC)**: 프로세스 간 데이터 공유 및 통신이 복잡함. 🔗🔴



 * **멀티스레딩 (Multithreading)**: **하나의 프로세스 내에서 여러 개의 스레드**를 동시에 실행하는 방식.

     * **장점**:

         * **낮은 오버헤드**: 스레드 생성 및 컨텍스트 스위칭 비용이 프로세스보다 훨씬 작음. ⚡️🔵

         * **쉬운 자원 공유 및 통신**: 메모리를 공유하므로 데이터 주고받기 용이. ✅🔵

         * **응답성 향상**: 시간이 오래 걸리는 작업을 백그라운드 스레드로 처리하고, 메인 스레드는 사용자 인터페이스 등을 계속 처리하여 응답성 유지 가능 (특히 GUI 앱).

     * **단점**:

         * **동기화 문제**: 여러 스레드가 공유 자원에 동시에 접근할 때 **경쟁 상태(Race Condition)** 나 **교착 상태(Deadlock)** 등 문제 발생 가능성. 이를 막기 위한 **동기화(Synchronization)** 메커니즘(뮤텍스, 세마포어 등) 필요. 복잡함! 🤯🔴

         * **안정성**: 한 스레드의 문제가 프로세스 전체에 영향을 줄 수 있음. 💥🔴

         * **(CPython 한정) GIL 문제**: CPU 바운드 작업에서 진정한 병렬 처리 효과 얻기 어려움 (아래 상세).



 * **Python 코드 예시 (개념)**: `multiprocessing`과 `threading` 모듈로 간단한 작업 실행.

 ### 🔒 파이썬 GIL

 1. **GIL (Global Interpreter Lock)이란?**

     - **CPython**의 잠금(Mutex) 🔒. **한 번에 딱 한 스레드**만 파이썬 코드를 실행하게 함.

     - **목적:** 메모리 관리(참조 카운팅)를 쉽게 스레드로부터 보호하기 위함.

 2. **핵심 영향 (`threading` 사용 시):**

     - **I/O 작업 (대기 많음):** 스레드가 대기할 때 다른 스레드가 일할 수 있어 **성능 향상 (동시성)** 👍🔵.

     - **CPU 작업 (계산 많음):** 여러 스레드도 결국 순서대로 실행되어 **병렬 처리 효과 거의 없음** 🐢🔴.

 3. **CPU 병렬 처리 해결책:**

     - ✅ **표준/안정적:** **`multiprocessing` 사용**. 각 프로세스가 독립된 GIL을 가져 **진짜 병렬 처리 가능** 🚀.

     - ✨ **실험적 (Python 3.13+):** **Free-threaded 모드**. GIL을 끄는 옵션으로 `threading` 병렬 가능성 제시.

 4. **⚠️ 중요 (Free-threaded):**

     - 아직 **매우 실험적인 기능**! (버그, 성능 저하, 호환성 문제 가능)

     - 기본 파이썬은 **여전히 GIL이 활성화**되어 있음.

````python
# %%
import time
import threading        # 멀티스레딩을 위한 모듈
import multiprocessing  # 멀티프로세싱을 위한 모듈
import os               # CPU 코어 수를 확인하기 위한 모듈
import random           # I/O 작업 시뮬레이션을 위한 랜덤 시간 생성

print(f"이 Colab 환경의 CPU 코어 수: {os.cpu_count()}")

# 실행할 작업의 수 (CPU 코어 수와 비슷하게 설정하면 비교하기 좋습니다)
NUM_TASKS = os.cpu_count() if os.cpu_count() else 2 # Colab에서 코어 수 반환 안될 경우 대비

# --- 1. I/O 바운드 작업 (대기 시뮬레이션) ---
def io_bound_task(task_id):
    """
    네트워크 요청이나 파일 읽기/쓰기처럼
    대기 시간이 긴 작업을 시뮬레이션합니다.
    """
    # 0.5초에서 1.0초 사이의 랜덤한 시간 동안 대기
    wait_time = random.uniform(0.5, 1.0)
    print(f"[I/O 작업 {task_id}] 시작 - {wait_time:.2f}초 동안 대기...")
    time.sleep(wait_time) # time.sleep()으로 대기 상태를 만듭니다.
    print(f"[I/O 작업 {task_id}] 완료!")

# --- 2. CPU 바운드 작업 (계산 시뮬레이션) ---
def cpu_bound_task(task_id):
    """
    복잡한 수학 계산처럼 CPU를 많이 사용하는 작업을 시뮬레이션합니다.
    """
    print(f"[CPU 작업 {task_id}] 시작 - 계산 중...")
    # 간단한 계산을 많이 반복하여 CPU 사용
    count = 0
    # 약 0.5초 동안 계산 반복 (시간은 조절 가능)
    end_time = time.time() + 0.5
    while time.time() < end_time:
        count += 1 # 간단한 연산 반복
    print(f"[CPU 작업 {task_id}] 완료 (반복 {count}회)")

# --- 실행 시간 측정 및 비교를 위한 함수들 ---

# 1. 순차 실행 함수
def run_sequentially(task_function, num_tasks):
    print(f"\n--- {task_function.__name__} 순차 실행 ({num_tasks}개 작업) ---")
    start_time = time.time()
    for i in range(num_tasks):
        task_function(i) # 작업을 하나씩 순서대로 실행
    end_time = time.time()
    duration = end_time - start_time
    print(f"--- 순차 실행 완료 ---")
    print(f"총 소요 시간: {duration:.4f} 초")
    return duration

# 2. 멀티스레딩 실행 함수
def run_with_threading(task_function, num_tasks):
    print(f"\n--- {task_function.__name__} 멀티스레딩 실행 ({num_tasks}개 스레드) ---")
    threads = []
    start_time = time.time()
    # 스레드 생성 및 시작
    for i in range(num_tasks):
        thread = threading.Thread(target=task_function, args=(i,))
        threads.append(thread)
        thread.start() # 스레드 실행 시작
    # 모든 스레드가 끝날 때까지 기다림
    for thread in threads:
        thread.join()
    end_time = time.time()
    duration = end_time - start_time
    print(f"--- 멀티스레딩 실행 완료 ---")
    print(f"총 소요 시간: {duration:.4f} 초")
    return duration

# 3. 멀티프로세싱 실행 함수
def run_with_multiprocessing(task_function, num_tasks):
    # 중요: Colab이나 특정 환경에서는 multiprocessing 사용 시
    #       if __name__ == '__main__': 구문이 필요할 수 있으나,
    #       노트북 셀에서는 보통 없어도 동작합니다.
    print(f"\n--- {task_function.__name__} 멀티프로세싱 실행 ({num_tasks}개 프로세스) ---")
    processes = []
    start_time = time.time()
    # 프로세스 생성 및 시작
    for i in range(num_tasks):
        process = multiprocessing.Process(target=task_function, args=(i,))
        processes.append(process)
        process.start() # 프로세스 실행 시작
    # 모든 프로세스가 끝날 때까지 기다림
    for process in processes:
        process.join()
    end_time = time.time()
    duration = end_time - start_time
    print(f"--- 멀티프로세싱 실행 완료 ---")
    print(f"총 소요 시간: {duration:.4f} 초")
    return duration

print("="*20, "I/O 바운드 작업 테스트", "="*20)

# 1. I/O 작업을 순차적으로 실행
seq_io_time = run_sequentially(io_bound_task, NUM_TASKS)

# 2. I/O 작업을 멀티스레딩으로 실행
thread_io_time = run_with_threading(io_bound_task, NUM_TASKS)

# 3. I/O 작업을 멀티프로세싱으로 실행 (참고용)
#    (I/O 작업에는 스레딩보다 오버헤드가 클 수 있음)
process_io_time = run_with_multiprocessing(io_bound_task, NUM_TASKS)

print("\n--- I/O 작업 결과 요약 ---")
print(f"순차 실행 시간: {seq_io_time:.4f} 초")
print(f"스레딩 실행 시간: {thread_io_time:.4f} 초")
print(f"프로세싱 실행 시간: {process_io_time:.4f} 초")
print("\n결론: I/O 바운드 작업은 스레드가 대기하는 동안 다른 스레드가 실행될 수 있으므로,")
print("      멀티스레딩이 순차 실행보다 훨씬 빠릅니다.")
print("      (멀티프로세싱도 가능하지만, 스레딩보다 오버헤드가 커서 덜 효율적일 수 있습니다.)")

print("\n\n", "="*20, "CPU 바운드 작업 테스트", "="*20)

# 1. CPU 작업을 순차적으로 실행
seq_cpu_time = run_sequentially(cpu_bound_task, NUM_TASKS)

# 2. CPU 작업을 멀티스레딩으로 실행
thread_cpu_time = run_with_threading(cpu_bound_task, NUM_TASKS)

# 3. CPU 작업을 멀티프로세싱으로 실행
process_cpu_time = run_with_multiprocessing(cpu_bound_task, NUM_TASKS)

print("\n--- CPU 작업 결과 요약 ---")
print(f"순차 실행 시간: {seq_cpu_time:.4f} 초")
print(f"스레딩 실행 시간: {thread_cpu_time:.4f} 초")
print(f"프로세싱 실행 시간: {process_cpu_time:.4f} 초")
print("\n결론: CPU 바운드 작업은 파이썬의 GIL 때문에 멀티스레딩으로 속도 향상을 보기 어렵습니다.")
print("      오히려 스레드 전환 비용 때문에 순차 실행보다 느릴 수도 있습니다.")
print("      반면, 멀티프로세싱은 여러 CPU 코어를 동시에 활용하므로 순차 실행보다 훨씬 빠릅니다.")
````

 ## 🐍✨ 파이썬에서의 효과적인 병렬/동시성 처리 방법 선택



 작업의 종류에 따라 적합한 모듈/방식을 선택하는 것이 중요합니다.



 * **I/O 바운드 작업 (네트워크, 파일 입출력 등 대기 시간 많음)**:

     * **`threading`**: 여러 작업을 동시에 시작하고 I/O 대기 시간에 다른 작업을 처리하여 전체 응답 시간 단축. GIL 영향 적음. 구현 비교적 간단.

     * **`asyncio`**: 비동기(Asynchronous) 프로그래밍 방식. 단일 스레드 내에서 이벤트 루프를 통해 여러 I/O 작업을 효율적으로 관리. `async`/`await` 문법 필요. 고성능 네트워크 애플리케이션 등에 유리. (학습 곡선 있음)



 * **CPU 바운드 작업 (계산 위주)**:

     * **`multiprocessing`**: 여러 개의 **독립적인 프로세스**를 생성하여 각 프로세스가 **별도의 GIL**을 가지고 **병렬적으로** CPU 코어를 활용. 진정한 병렬 처리 가능. 단, 프로세스 간 통신(IPC) 필요 시 복잡성 증가 및 오버헤드 발생.



 * **GPU 가속 필요한 작업 (딥러닝 등)**:

     * **전문 라이브러리 활용**: TensorFlow, PyTorch, CuPy 등 GPU 연산을 지원하는 라이브러리 사용. 이 라이브러리들이 내부적으로 GPU 병렬 처리를 관리해 줌.



 **결론**: 작업의 병목 지점이 CPU인지 I/O인지 파악하고, 그에 맞는 동시성/병렬 처리 방식을 선택하는 것이 파이썬 프로그램의 성능을 최적화하는 핵심입니다.

 ## 🚦⏱️ 스케줄링 (Scheduling)



 * **정의**: 운영체제가 시스템의 자원(특히 **CPU**)을 여러 프로세스 또는 스레드에게 **어떻게 할당하고 관리**할 것인지 결정하는 정책 및 과정.

 * **목적**:

     * **공정성(Fairness)**: 모든 프로세스/스레드가 공평하게 자원 할당받도록.

     * **효율성(Efficiency)**: CPU 등 자원 활용률 극대화.

     * **응답 시간(Response Time) 최소화**: 사용자 요청에 빠르게 반응.

     * **처리율(Throughput) 최대화**: 단위 시간당 처리하는 작업 수 최대화.

     * **대기 시간(Waiting Time) 최소화**: 프로세스/스레드가 준비 큐에서 기다리는 시간 최소화.

 * **스케줄러(Scheduler)**: 스케줄링 정책을 실제로 수행하는 운영체제 커널의 일부.

 * **기본적인 스케줄링 알고리즘 (이름만 소개)**:

     * **FCFS (First-Come, First-Served)**: 먼저 온 순서대로 처리. (비선점형)

     * **SJF (Shortest Job First)**: 실행 시간이 가장 짧은 작업 먼저 처리. (평균 대기 시간 최소화, 기아(Starvation) 문제 가능)

     * **Round Robin (RR)**: 각 작업에 동일한 시간 할당량(Time Slice/Quantum)을 주고, 시간 내 못 끝내면 맨 뒤로 보내는 방식. (시분할 시스템, 선점형)

     * **Priority Scheduling**: 우선순위가 높은 작업 먼저 처리. (기아 문제 가능)

     * (이 외에도 Multilevel Queue, Multilevel Feedback Queue 등 다양한 알고리즘 존재)

 ## ❓ 객관식, 단답형 및 서술형 문제



 * 각 문제의 난이도를 ⭐️ ~ ⭐️⭐️⭐️⭐️⭐️ 로 표시했습니다.

 ### 🧐 객관식 문제 1 (난이도: ⭐️⭐️)



 운영체제(OS)의 핵심 기능으로 보기 **어려운** 것은 무엇인가요?



 1.  프로세스 관리 및 스케줄링

 2.  메모리 할당 및 관리

 3.  웹 브라우저 기능 제공

 4.  파일 시스템 관리

 5.  입출력 장치 제어

 ### 🧐 객관식 문제 2 (난이도: ⭐️⭐️⭐️)



 **프로세스(Process)** 와 **스레드(Thread)** 의 관계에 대한 설명으로 **가장 올바른** 것은 무엇인가요?



 1.  하나의 스레드는 여러 개의 프로세스를 포함할 수 있다.

 2.  프로세스와 스레드는 완전히 동일한 개념이다.

 3.  하나의 프로세스는 반드시 하나의 스레드만 가질 수 있다.

 4.  스레드는 프로세스 내에서 실행되는 단위이며, 한 프로세스는 여러 스레드를 가질 수 있다.

 5.  프로세스는 스레드의 자원(메모리 등)을 공유하여 사용한다.

 ### 🧐 객관식 문제 3 (난이도: ⭐️⭐️⭐️⭐️)



 파이썬(CPython)의 **GIL(Global Interpreter Lock)** 에 대한 설명으로 **틀린** 것은 무엇인가요?



 1.  전역 인터프리터 락의 약자이다.

 2.  한 번에 하나의 스레드만 파이썬 바이트코드를 실행하도록 제어한다.

 3.  CPU 바운드 작업의 멀티스레딩 성능을 크게 향상시킨다.

 4.  I/O 바운드 작업에서는 스레드가 대기하는 동안 다른 스레드가 GIL을 얻어 실행될 수 있다.

 5.  멀티프로세싱(`multiprocessing`) 방식은 GIL의 영향을 받지 않는다.

 ### ✍️ 단답형 / 서술형 문제 1 (난이도: ⭐️⭐️⭐️⭐️)



 **멀티프로세싱(Multiprocessing)** 과 **멀티스레딩(Multithreading)** 의 **장점**과 **단점**을 각각 한 가지 이상 비교 설명하세요.

 ### ✍️ 단답형 / 서술형 문제 2 (난이도: ⭐️⭐️⭐️⭐️⭐️)



 파이썬에서 **CPU 연산 위주의 작업**(예: 큰 행렬 곱셈 계산)을 **병렬로 처리**하여 성능을 높이고 싶을 때, `threading` 모듈 대신 **`multiprocessing` 모듈을 사용해야 하는 이유**를 **GIL(Global Interpreter Lock)** 과 관련지어 설명하세요.

## 💭 마무리



 오늘 3교시에는 컴퓨터 시스템의 핵심 관리자인 **운영체제(OS)** 의 역할과 기능을 이해하고, 프로그램 실행의 기본 단위인 **프로세스**와 **스레드**의 개념 및 중요한 차이점을 배웠습니다.



 **멀티프로세싱**과 **멀티스레딩**을 통해 여러 작업을 동시에 처리하는 방법과 각각의 장단점을 비교했으며, CPU 자원 분배를 위한 **스케줄링**의 기본 원리도 살펴보았습니다.



 특히 파이썬(CPython) 환경에서 멀티스레딩 사용 시 주의해야 할 **GIL(Global Interpreter Lock)** 의 개념과 이것이 **CPU 바운드/I/O 바운드 작업 성능**에 미치는 영향을 이해하는 것은 매우 중요합니다. 이를 바탕으로 작업 유형에 따라 **`threading`과 `multiprocessing` 중 어떤 방식을 선택**하는 것이 효과적인지 판단하는 기준을 세웠습니다.
